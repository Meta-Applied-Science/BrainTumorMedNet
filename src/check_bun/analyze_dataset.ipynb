## Check Image Channel and resolution

```python
import os
import cv2

# Root folder containing subfolders
root_folder = '/home/ma012/AlexServer/phase2/dataset/BT_Large_4c/'  # Adjust to your path

# Valid image extensions
valid_extensions = ('.jpg')

unique_resolution = []

# Walk through all subdirectories
for dirpath, _, filenames in os.walk(root_folder):
    for filename in filenames:
        if filename.lower().endswith(valid_extensions):
            file_path = os.path.join(dirpath, filename)
            
            # Read image
            image = cv2.imread(file_path)
            if image is None:
                print(f"Failed to load: {file_path}")
                continue
            
            height, width = image.shape[:2]

            resolution = f"{width}x{height}"

            if(resolution not in unique_resolution):
                unique_resolution.append(resolution)

            channels = image.shape[2] if len(image.shape) == 3 else 1

            print(f"{file_path} - Resolution: {width}x{height}, Channels: {channels}")
```

## Check for Balanced Number of Images

```python
import os
from collections import defaultdict

# Root folder path
root_folder =  '/home/ma012/AlexServer/phase2/dataset/BT_Large_4c/'  # Adjust as needed

# Valid image extensions
valid_extensions = ('.jpg')

# Save total images of each folder
folder_total_images = defaultdict(int)

# Mapping images to closest parent folder
file_to_folder = {}

# Scan all sub folders
for dirpath, _, filenames in os.walk(root_folder):
    for filename in filenames:
        if filename.lower().endswith(valid_extensions):
            full_path = os.path.join(dirpath, filename)
            
            # Up number of images of parent folder
            subpath = dirpath
            while subpath.startswith(root_folder):
                folder_total_images[subpath] += 1
                parent = os.path.dirname(subpath)
                if parent == subpath:  # root reached
                    break
                subpath = parent

# show results
for folder, count in sorted(folder_total_images.items()):
    print(f"{folder}: {count} images")
```
